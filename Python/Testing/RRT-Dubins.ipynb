```python
env = StaticEnvironment((200, 200), 8)
env.plot()
```

```python
for obs in env.obstacles:
    print(obs.points)
```

```python
rrt = RRT(env)

start = env.random_free_space()
end = env.random_free_space()
rrt.set_start(start)
rrt.run(end, nb_iteration=100)

env.plot()
rrt.plot(nodes=True)
```
